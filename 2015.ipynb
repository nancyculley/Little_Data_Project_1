```python
# Dependencies and Setup
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from scipy import stats

#throw in a "hurricane" line for fun even though there weren't any hurricanes in 2015.
```

```python
# Load Hurricane Data
hurricanes = "resources/Hurricane_data_all.csv"

# Load Index data
sp1500 = "resources/S&P1500_Sector_All.csv"
```

```python
# Read the Index and Hurricane csv
sp1500 = pd.read_csv(sp1500)
hurricanes = pd.read_csv(hurricanes)
```

```python
#rename columns so they'll be easier to call out later

sp1500.columns = ['Date', 'Consumer_Discretionary', 'Communication_Services', 'Consumer_Staples', 
                      'Energy', 'Financials', 'Health_Care', 'Industrials', 'Information_Technology',
                      'Materials', 'Real_Estate', 'Utilities']
sp1500.head()
```

```python
#split the data to only focus on just S&P stocks from 2015
#June 2015 data

sp1500['Date'] = pd.to_datetime(sp1500['Date'])

sp1500_june = sp1500[(sp1500['Date'] > '6/1/2015') & (sp1500['Date'] <= '6/30/2015')]
sp1500_june = sp1500_june.reset_index(drop = True)
```

```python
#hurricanes
```

```python
#June 2015 graph

for x in range(1, 12):
   plt.plot(sp1500_june['Date'], sp1500_june.iloc[:,x])
plt.legend()
plt.ylabel('Total Return')
plt.xlabel('Date (2015)')
plt.title('Change in Total Return over June 2015')
plt.legend(loc='center left', bbox_to_anchor=(1.0, 0.5))
plt.xticks(rotation = 45)
plt.show()
```
